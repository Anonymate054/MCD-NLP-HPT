```python
import scripts.yt_utils as yt_utils
import pandas as pd
import json
import fs
```

```python
INTERIM_VIDEO_COMMETS_DIR = fs.open_fs("../../data/interim/video_commets")
csv_files = [file for file in INTERIM_VIDEO_COMMETS_DIR.listdir('/') if file.endswith('.csv')]
csv_files
```

```python
DIR_CSV = INTERIM_VIDEO_COMMETS_DIR.getsyspath(csv_files[0])
```

```python
df = pd.read_csv(DIR_CSV)
df.head()
```

```python
df_replies = df[df["replies"] != "{}"]
df_replies
```

```python
(df_replies["replies"])
```

```python
# Extraer los comentarios
comments_list = df_replies["replies"].apply(lambda x: x['comments'] if isinstance(x, dict) and 'comments' in x else [])

# Crear un DataFrame a partir de la lista de comentarios
comments_df = comments_list.explode().reset_index(drop=True)

# Extraer el texto de los comentarios, si hay comentarios
if not comments_df.empty:
    comments_df = comments_df.apply(lambda x: x['text'] if isinstance(x, dict) and 'text' in x else None)

# Para ver el resultado
print(comments_df)
```
