```python
import scripts.yt_utils as yt_utils
import scripts.dir_utils as dir_utils
import pandas as pd
import json
import fs
import re
```

```python
PROCESSED_FILES_DIR = fs.open_fs("../../data/processed")
PROCESSED_COMMENTS_COMBINED_DIR = PROCESSED_FILES_DIR.getsyspath("Combined.csv")
PROCESSED_COMMENTS_REPLIES_DIR = PROCESSED_FILES_DIR.getsyspath("Replies.csv")
PROCESSED_COMMENTS_ITEMS_DIR = PROCESSED_FILES_DIR.getsyspath("Items.csv")
```

```python
INTERIM_PARSED_FILES_DIR = fs.open_fs("../../data/interim/parsed_files")
```

```python
INTERIM_VIDEO_REQUEST_DIR = fs.open_fs("../../data/interim/video_request")
request_files = [file for file in INTERIM_VIDEO_REQUEST_DIR.listdir('/') if file.endswith('.json')]
request_files
```

```python
df_full_combined = pd.DataFrame()
df_full_replies = pd.DataFrame()
df_full_items = pd.DataFrame()

for request in request_files:
    DIR_JSON = INTERIM_VIDEO_REQUEST_DIR.getsyspath(request)
    VideoID_match = re.search(r'video_(.+?)_request\.json', request)
    if VideoID_match:  
        VideoID = VideoID_match.group(1)
        print(f"Processing video: {VideoID}")

    with open(DIR_JSON, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    df_combined, df_replies, df_items = yt_utils.parse_comments_v2(data)

    dir_utils.mkdir(INTERIM_PARSED_FILES_DIR, VideoID)
    INTERIM_PARSED_FILES_VIDEOID_DIR = fs.open_fs(f"../../data/interim/parsed_files/{VideoID}")
    PARSED_COMMENTS_COMBINED_DIR = INTERIM_PARSED_FILES_VIDEOID_DIR.getsyspath(f"{VideoID}_combined.csv")
    PARSED_COMMENTS_REPLIES_DIR = INTERIM_PARSED_FILES_VIDEOID_DIR.getsyspath(f"{VideoID}_replies.csv")
    PARSED_COMMENTS_ITEMS_DIR = INTERIM_PARSED_FILES_VIDEOID_DIR.getsyspath(f"{VideoID}_items.csv")

    df_combined.to_csv(PARSED_COMMENTS_COMBINED_DIR, encoding='utf-8', index=False)
    df_replies.to_csv(PARSED_COMMENTS_REPLIES_DIR, encoding='utf-8', index=False)
    df_items.to_csv(PARSED_COMMENTS_ITEMS_DIR, encoding='utf-8', index=False)

    df_full_combined = pd.concat([df_full_combined, df_combined], ignore_index=True)
    df_full_replies = pd.concat([df_full_replies, df_replies], ignore_index=True)
    df_full_items = pd.concat([df_full_items, df_items], ignore_index=True)


df_full_combined.to_csv(PROCESSED_COMMENTS_COMBINED_DIR, encoding='utf-8', index=False)
df_full_replies.to_csv(PROCESSED_COMMENTS_REPLIES_DIR, encoding='utf-8', index=False)
df_full_items.to_csv(PROCESSED_COMMENTS_ITEMS_DIR, encoding='utf-8', index=False)
```
