# Libraries

```python
import googleapiclient.discovery
from dotenv import load_dotenv
import pandas as pd
import os
import fs
```

```python
INTERIM_PROJECT_DIR = fs.open_fs("../../data/interim")
YT_VIDEOS_CSV = INTERIM_PROJECT_DIR.getsyspath("YT_videos.csv")
```

Read enviromental variables

```python
load_dotenv()

YT_API_KEY = os.getenv('YT-API-KEY')
```

## Pull all videos

```python
channelId = "UCeYSnd34IPe57sM8hkjWhrw"
api_service_name = "youtube"
api_version = "v3"

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=YT_API_KEY)

request = youtube.playlists().list(
    part="snippet",
    channelId=channelId,
    maxResults=100
)

# Execute the request.
response = request.execute()
response
```

```python
playlistId = response['items'][0]["id"]
playlistId
```

```python
api_service_name = "youtube"
api_version = "v3"

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=YT_API_KEY)

request = youtube.playlistItems().list(
    part="snippet",
    playlistId=playlistId,
    maxResults=100
)

# Execute the request.
response = request.execute()
response
```

```python
#  Extract snippet and videoId
data = []
for item in response['items']:
    snippet = item['snippet']
    video_id = snippet['resourceId']['videoId']
    # Remove 'thumbnails' and flatten 'resourceId' to include 'videoId'
    snippet_data = {
        'publishedAt': snippet['publishedAt'],
        'channelId': snippet['channelId'],
        'title': snippet['title'],
        'description': snippet['description'],
        'channelTitle': snippet['channelTitle'],
        'playlistId': snippet['playlistId'],
        'position': snippet['position'],
        'videoId': video_id,
        'videoOwnerChannelTitle': snippet['videoOwnerChannelTitle'],
        'videoOwnerChannelId': snippet['videoOwnerChannelId']
    }
    data.append(snippet_data)

# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
df.to_csv(YT_VIDEOS_CSV, encoding='utf-8', index=False)
```
