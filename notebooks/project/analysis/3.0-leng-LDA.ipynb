```
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.probability import FreqDist
from nltk.util import ngrams
import pandas as pd

from gensim import corpora
import gensim

import ast
import fs
```

```
INTERIM_FILES_DIR = fs.open_fs("MCD-NLP-HPT/data/interim")
INTERIM_COMMENTS_CLEANED_BY_VIDEO_DIR = INTERIM_FILES_DIR.getsyspath("Comments_cleaned_by_video.csv")
```

```
df_video_corpus = pd.read_csv(INTERIM_COMMENTS_CLEANED_BY_VIDEO_DIR)

# Convert to list
df_video_corpus['Tokens_without_stopwords_lemmatized_combined'] = df_video_corpus['Tokens_without_stopwords_lemmatized_combined'].apply(ast.literal_eval)
df_video_corpus['Tokens_combined'] = df_video_corpus['Tokens_combined'].apply(ast.literal_eval)
df_video_corpus['List_of_Tokens_combined'] = df_video_corpus['List_of_Tokens_combined'].apply(ast.literal_eval)
df_video_corpus['List_of_WSL_Tokens_combined'] = df_video_corpus['List_of_WSL_Tokens_combined'].apply(ast.literal_eval)
df_video_corpus.head(5)
```

```
# df_video_corpus['List_of_Tokens_combined'][0]
```

```
# text = ["The movie was very scary and long.", "The Movie was not scary,","The movie was good"]
corpus = df_video_corpus['List_of_WSL_Tokens_combined'][1]
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpus)
features = vectorizer.get_feature_names_out()
dense = vectors.todense()
denselist = dense.tolist()
df_dense = pd.DataFrame(denselist, columns=features)
df_dense
```

```
Top = 4
Category = pd.DataFrame()
def Get_Category_words_TOP(Top,Panda):
    Panda = Panda.T.sort_index(axis=1, ascending=True)
    Arr_words = []
    shape = Panda.shape
    Panda = Panda.reset_index()
    #print(shape[1]-1)
    for j in range(shape[1]):
        #print(j,'-------')
        Panda = Panda.reset_index()
        Panda.drop('level_0', axis=1, inplace=True)
        Panda.sort_values(by=[j],inplace=True, ascending=False)
        Panda = Panda.reset_index()
        Panda.drop('level_0', axis=1, inplace=True)
        for i in range (Top):
            Arr_words.append(Panda['index'][i])
        Category[j] = [Arr_words]
        Arr_words = []
    return Category
```

```
Category_Words = Get_Category_words_TOP(Top,df_dense)
Category_Words
```

```
from gensim import corpora
documents = [comment.split() for comment in df_video_corpus['List_of_WSL_Tokens_combined'][4]]
from gensim import corpora

# Crear el diccionario
dictionary = corpora.Dictionary(documents)

# Ver el diccionario (opcional)
print(dictionary.token2id)  # Muestra la correspondencia palabra-ID
```

```
# documents
```

```
# Crear el corpus
corpus = [dictionary.doc2bow(doc) for doc in documents]

# Ver el corpus (opcional)
print(corpus)  # Ejemplo: [(0, 1), (1, 1)] -> palabras con su frecuencia
```

```
from gensim.models import LdaModel

# Crear el modelo LDA
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=10)

# Ver los tópicos generados
topics = lda_model.print_topics(num_words=3)  # Muestra las 4 palabras más representativas de cada tópico
for topic in topics:
    print(topic)
```
