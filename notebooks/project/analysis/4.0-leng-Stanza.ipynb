```
import stanza

# Configurar el pipeline de Stanza
nlp = stanza.Pipeline(lang="es", processors="tokenize,mwt,pos,lemma")

# Texto a analizar
text = "Los gatos están durmiendo tranquilamente en el sofá."

# Procesar el texto
doc = nlp(text)

# Iterar sobre las oraciones y tokens para mostrar los resultados
for i, sentence in enumerate(doc.sentences, 1):
    print(f"Oración {i}:")
    for word in sentence.words:
        print(f"Texto: {word.text}, Lema: {word.lemma}, POS: {word.upos}, XPOS: {word.xpos}")
```
