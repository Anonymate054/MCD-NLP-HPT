```
from nltk.util import ngrams
from nltk.probability import FreqDist
import pandas as pd
import ast
import fs
```

```
INTERIM_FILES_DIR = fs.open_fs("MCD-NLP-HPT/data/interim")
INTERIM_COMMENTS_CLEANED_BY_VIDEO_DIR = INTERIM_FILES_DIR.getsyspath("Comments_cleaned_by_video.csv")
```

```
df_video_corpus = pd.read_csv(INTERIM_COMMENTS_CLEANED_BY_VIDEO_DIR)

# Convert to list
df_video_corpus['Tokens_without_stopwords_lemmatized_combined'] = df_video_corpus['Tokens_without_stopwords_lemmatized_combined'].apply(ast.literal_eval)
df_video_corpus['Tokens_combined'] = df_video_corpus['Tokens_combined'].apply(ast.literal_eval)

# Show information
df_video_corpus.head()
```

# Create N-grams

```
df_video_corpus['Tokens_combined_bigrams'] = df_video_corpus['Tokens_combined'].apply(lambda x: list(ngrams(x, 2)))
df_video_corpus['Tokens_without_stopwords_lemmatized_combined_bigrams'] = df_video_corpus['Tokens_without_stopwords_lemmatized_combined'].apply(lambda x: list(ngrams(x, 2)))
df_video_corpus['Tokens_combined_trigrams'] = df_video_corpus['Tokens_combined'].apply(lambda x: list(ngrams(x, 3)))
df_video_corpus['Tokens_without_stopwords_lemmatized_combined_trigrams'] = df_video_corpus['Tokens_without_stopwords_lemmatized_combined'].apply(lambda x: list(ngrams(x, 3)))
df_video_corpus['Tokens_WSL_bigrams_most_frec'] = df_video_corpus['Tokens_without_stopwords_lemmatized_combined_bigrams'].apply(lambda x: FreqDist(x).most_common(10))
df_video_corpus['Tokens_WSL_trigrams_most_frec'] = df_video_corpus['Tokens_without_stopwords_lemmatized_combined_trigrams'].apply(lambda x: FreqDist(x).most_common(10))
df_video_corpus
```
